```python

from pdfkb.loader import load_multimodal
from pdfkb.chunker import chunk_text
from pdfkb.summarizer import summarize_for_teaching

pdfs = ["paper.pdf"]

content = load_multimodal(pdfs)
chunks = chunk_text(content)

teaching_summary = summarize_for_teaching(chunks, model="llama3")
print(teaching_summary)
```
